```python
#4-1
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
import pandas as pd


```

```python
titanic_df = sns.load_dataset("titanic")
display(titanic_df)
```

```python
#4-2-0 IMPORT DATA
dataset =titanic_df

a1=pd.DataFrame()
a1["Number"]=dataset.isnull().sum()
df1 = pd.DataFrame({"name":["survived","pclass","sex","age","sibsp","parch","fare","embarked","class","who","adult_male","deck","embark_town","alive","alone"],"number":[0,0,0,177,0,0,0,2,0,0,0,688,2,0,0]})
a2=pd.DataFrame()
a2["Number"]=dataset.isnull().sum(axis=1)

```

```python
#4-2-a 
fig=plt.figure(figsize=(30,10))
ax1 = fig.add_subplot()
sns.catplot(x="number", y="name", data=df1, kind="bar",ax=ax1)
```

```python

#4-2-b 
fig2=plt.figure(figsize=(10,10))
ax2 = fig2.add_subplot()
sns.catplot(x="Number", data=a2, kind="count",ax=ax2)

```

```python
#4-2-c 
fig3=plt.figure(figsize=(10,10))
ax3 = fig3.add_subplot()
sns.catplot(x="number", y="name", data=df1, kind="bar",ax=ax3)
sns.catplot(x="Number", data=a2, kind="count",ax=ax3)
```
